```python
## timer 메서드 가져오기
import time
def timer(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        computation_time = end_time - start_time
        print(f"Execution time of {func.__name__}: {computation_time} seconds")
        return result
    return wrapper
```

```python
class cumstom_PBL_train_dataset(torch.utils.data.Dataset):
  def __init__(self, train=True, transform=None):

    ## 학습 모드: train이 True이면 train 데이터셋
    if train:
      self.data = train_data
    ## 검증 모드: train이 False이면 valid 데이터셋
    else:
      self.data = valid_data

    ## 레이블은 정보량이 많지 않으므로 target에 전체 레이블 저장
    self.target = np.array(self.data[['n','potential']])

    ## 특별한 변환이 없으면 그대로 tensor로 변환하도록 설정: Tensor class
    if transform is None:
      self.transform = torch.Tensor
    else:
      self.transform = transform

  def __len__(self):
    return len(self.target)

  def __getitem__(self, index):

    ## 위 셀에서 실험한 대로 self.data에 저장한 데이터프레임에서 index에 해당하는 이미지 경로 설정, 불러와서 array 형태로 저장
    image_path = '/content/drive/MyDrive/Colab Notebooks/data/pbl_data/' + self.data['img_path'][index]
    self.input = np.array(Image.open(image_path))

    ## transform에 지정한 변환을 거침: 아무것도 안 하는 경우 Tensor()를 통해 tensor로 변환
    img = self.transform(self.input)
    label = self.transform(self.target[index])

    return img, label
```

```python
## Dataset을 지정하고
custom_train_data = cumstom_PBL_train_dataset(train=True)
custom_test_data = cumstom_PBL_train_dataset(train=False)

min_batch = 100

## train_loader와 test_loader에 데이터 적용
train_loader = torch.utils.data.DataLoader(dataset=custom_train_data, batch_size=min_batch, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=custom_test_data, batch_size=min_batch, shuffle=False)
```

```python
class cumstom_PBL_train_dataset2(torch.utils.data.Dataset):
  def __init__(self, train=True, transform=None):

    if train:
      self.data = train_data
    else:
      self.data = valid_data

    ## 포텐셜에 대한 회귀 생략, 양자 개수만 레이블에 일단 입력
    self.target = np.array(self.data[['n']])

    if transform is None:
      self.transform = torch.Tensor
    else:
      self.transform = transform

  def __len__(self):
    return len(self.target)

  def __getitem__(self, index):

    image_path = '/content/drive/MyDrive/Colab Notebooks/data/pbl_data/' + self.data['img_path'][index]
    self.input = np.array(Image.open(image_path))

    img = self.transform(self.input)
    label = self.transform(self.target[index])

    return img, label
```

```python
## Dataset 재적용
custom_train_data = cumstom_PBL_train_dataset2(train=True)
custom_test_data = cumstom_PBL_train_dataset2(train=False)

min_batch = 100

train_loader = torch.utils.data.DataLoader(dataset=custom_train_data, batch_size=min_batch, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=custom_test_data, batch_size=min_batch, shuffle=False)
```

```python
## 3주차 예제에서 다룬 분류 모델 구조를 그대로 적용
class DeepClassification(nn.Module):
    def __init__(self, hidden_variables=[128, 64, 10], input_output_dim=(28*28, 10)):
        super().__init__()
        # Concise Implementation
        self.input_variable_dim = input_output_dim[0]
        self.output_variable_dim = input_output_dim[1]
        self.list_hidden_variable = hidden_variables
        self.layer = nn.Sequential()

        variable_dim = self.input_variable_dim
        for i, hidden_variable in enumerate(self.list_hidden_variable):
            self.layer.add_module('layer_' + str(i), nn.Linear(variable_dim, hidden_variable))
            self.layer.add_module('custom_activation_' + str(i), nn.ReLU())
            variable_dim = hidden_variable
        self.layer.add_module('final_layer', nn.Linear(variable_dim, self.output_variable_dim))

    def forward(self, x):
    # Computes the outputs / predictions
        x = x.view(-1, self.input_variable_dim)
        y_hat = self.layer(x)
        return y_hat
```

```python
## tr의
def make_train_step(model, loss_fn, optimizer):
    def train_step_fn(x, y):

        model.train()
        y_hat = model(x)
        print(torch.sum(torch.argmax(y_hat, axis=1) == torch.max(y, axis=1)[0].to(torch.long)))
        loss = loss_fn(y_hat, torch.max(y, axis=1)[0].to(torch.long))

        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        return loss.item()
    return train_step_fn
```

```python
@timer
def train_model_classification(epochs=1000, eval_test_accuracy=False):

    def evaluate_step(x, y):
      y_hat = model(x)
      print(y_hat)
      print(torch.argmax(y_hat, axis=1))
      print(torch.max(y, axis=1)[0].to(torch.long))
      result = torch.sum(torch.argmax(y_hat, axis=1) == torch.max(y, axis=1)[0].to(torch.long))
      print(result,"\n")
      return result, len(y)

    train_losses = []
    for epoch in range(epochs):
        mini_batch_losses = []
        step_time = []
        step_time.append(time.time())
        for x_minibatch, y_minibatch in train_loader:
            x_minibatch = x_minibatch.to(device)
            y_minibatch = y_minibatch.to(device)
            mini_batch_loss = train_step(x_minibatch, y_minibatch)
            mini_batch_losses.append(mini_batch_loss)
            step_time.append(time.time())
            print("step time: ", round(step_time[-1] - step_time[-2], 4) , ", loss:", round(mini_batch_loss, 4), ", epoch ",epoch, ", minibatch", len(mini_batch_losses), "\n")

        if (epoch + 1) % 1 == 0:
            loss = np.mean(mini_batch_losses)
            print("train loss at {} epoch:{}".format(epoch + 1, loss))
            train_losses.append(loss)

            if eval_test_accuracy:
                with torch.no_grad():
                    test_accuracy = 0
                    test_result = 0
                    test_cnt = 0
                    for x_minibatch_test, y_minibatch_test in test_loader:
                        x_minibatch_test = x_minibatch_test.to(device)
                        y_minibatch_test = y_minibatch_test.to(device)
                        result, cnt = evaluate_step(x_minibatch_test, y_minibatch_test)
                        test_result += result
                        test_cnt += cnt
                    test_accuracy = 100 * test_result / test_cnt
                    print("test accuracy at {} epoch: {}%".format(epoch + 1, test_accuracy))

            checkpoint = {'epochs': epochs,
                          'model_state_dict': model.state_dict(),
                          'optimizer_state_dict': optimizer.state_dict(),
                          'train_loss': train_losses,
                          'test_accuracy': test_accuracy
                          }
            torch.save(checkpoint, 'model_checkpoint.pth')


    return train_losses, test_accuracy
```

```python
model = DeepClassification(input_output_dim=(10*300, 9)).to(device)
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
train_step = make_train_step(model, loss_fn, optimizer)
```

```python
## batch를 600번 학습해야 한 epoch가 끝나는데, 한번 학습에 25초 내외가 소요됨 - 정상적인 학습 불가능
## 모델 학습이나 loss 계산에는 시간이 오래 걸리지 않으나, 이미지 load에서 시간이 크게 소요됨
## 마운트된 구글 드라이브 폴더 내에 과도하게 많은 파일이 있어 조회 작업에서 부하가 발생한 것으로 추정 - 따라서 부적절한 방식
train_loss, test_accuracy = train_model_classification(epochs=5, eval_test_accuracy=True)
```

```python
## validation accuracy 코드가 정상 작동되는 것을 확인하기 위해 next(iter(test_loader))로 한 batch만 예시로 뽑아 정확도 계산
def evaluate_step(x, y):
  y_hat = model(x)
  print(y_hat)
  print(torch.argmax(y_hat, axis=1))
  print(torch.max(y, axis=1)[0].to(torch.long))
  result = torch.sum(torch.argmax(y_hat, axis=1) == torch.max(y, axis=1)[0].to(torch.long))
  print(result,"\n")
  return result, len(y)

test_accuracy = 0
test_result = 0
test_cnt = 0
epoch = 0

x_minibatch_test, y_minibatch_test = next(iter(test_loader))
x_minibatch_test = x_minibatch_test.to(device)
y_minibatch_test = y_minibatch_test.to(device)
result, cnt = evaluate_step(x_minibatch_test, y_minibatch_test)
test_result += result
test_cnt += cnt
test_accuracy = 100 * test_result / test_cnt
print("test accuracy at {} epoch: {}%".format(epoch + 1, test_accuracy))
```
